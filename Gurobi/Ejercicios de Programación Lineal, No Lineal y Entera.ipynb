## Programación Lineal (LP)

image.png

```python
# Crear el modelo
model = Model("Ejemplo_Maximizacion_Lineal")

# Variables de decisión (enteras y no negativas)
x1 = model.addVar(vtype=GRB.CONTINUOUS, lb=0, name="x1")
x2 = model.addVar(vtype=GRB.CONTINUOUS, lb=0, name="x2")

# Función objetivo: Max 3x1 + 5x2
model.setObjective(3*x1 + 5*x2, GRB.MAXIMIZE)

# Restricciones
model.addConstr(x1 + x2 <= 16, name="R1")
model.addConstr(x2 <= 7.5, name="R2")

# Optimizar
model.optimize()

# Mostrar resultados
if model.Status == GRB.OPTIMAL:
    print("\nStatus: Óptimo")
    print(f"x1 = {x1.X}")
    print(f"x2 = {x2.X}")
    print(f"Valor óptimo (Z*) = {model.ObjVal}")
else:
    print("No se encontró solución óptima.")
```

```python
# Crear el modelo
model = Model("Ejemplo_Maximizacion_Lineal")

# Variables de decisión (enteras y no negativas)
x1 = model.addVar(vtype=GRB.CONTINUOUS, lb=0, name="x1")
x2 = model.addVar(vtype=GRB.CONTINUOUS, lb=0, name="x2")

# Función objetivo: Max 3x1 + 5x2
model.setObjective(x1 + 3*x2, GRB.MAXIMIZE)

# Restricciones
model.addConstr(-x1 + x2 <= 3, name="R1")
model.addConstr(-x1 + 2*x2 <= 8, name="R2")
model.addConstr(3*x1 + x2 <= 18, name="R3")

# Optimizar
model.optimize()

# Mostrar resultados
if model.Status == GRB.OPTIMAL:
    print("\nStatus: Óptimo")
    print(f"x1 = {x1.X}")
    print(f"x2 = {x2.X}")
    print(f"Valor óptimo (Z*) = {model.ObjVal}")
else:
    print("No se encontró solución óptima.")
```

## Programación Entera

```python
# Crear el modelo
model = Model("Ejemplo_Maximizacion_Entera")

# Variables de decisión (enteras y no negativas)
x1 = model.addVar(vtype=GRB.INTEGER, lb=0, name="x1")
x2 = model.addVar(vtype=GRB.INTEGER, lb=0, name="x2")

# Función objetivo: Max 3x1 + 5x2
model.setObjective(3*x1 + 5*x2, GRB.MAXIMIZE)

# Restricciones
model.addConstr(x1 + x2 <= 16, name="R1")
model.addConstr(x2 <= 7.5, name="R2")

# Optimizar
model.optimize()

# Mostrar resultados
if model.Status == GRB.OPTIMAL:
    print("\nStatus: Óptimo")
    print(f"x1 = {x1.X}")
    print(f"x2 = {x2.X}")
    print(f"Valor óptimo (Z*) = {model.ObjVal}")
else:
    print("No se encontró solución óptima.")
```

## Programación No Lineal (NLP)

image.png

```python
# Crear la instancia del modelo
model = Model("Minimizar_Coeficientes")

# Inicialización de las variables de decisión
B1 = model.addVar(vtype=GRB.CONTINUOUS, lb=-GRB.INFINITY, ub=GRB.INFINITY, name="B1")
B2 = model.addVar(vtype=GRB.CONTINUOUS, lb=-GRB.INFINITY, ub=GRB.INFINITY, name="B2")
a  = model.addVar(vtype=GRB.CONTINUOUS, lb=-GRB.INFINITY, ub=GRB.INFINITY, name="a")

# Agregar la función objetivo
model.setObjective(1/2 * (B1 ** 2 + B2 ** 2), GRB.MINIMIZE)

# Agregar las restricciones

model.addConstr(a + 4*B1 + 2*B2 >= 1, "Positivo")
model.addConstr(a + 2*B1 + B2 >= 1, "Positivo")
model.addConstr(a + B1 + 3*B2 <= -1, "Negativo")

# Resolver el problema
model.optimize()

# -----------------------------
# Resultados
# -----------------------------
if model.Status == GRB.OPTIMAL:
    print("Solución óptima:")

    # Resultados
    for var in model.getVars():
      print(var.VarName, "=", var.X)

    print(f"\nFunción Objetivo = {model.objVal:.2f}")
else:
    print("No se encontró solución óptima (modelo infeasible o ilimitado).")
```
