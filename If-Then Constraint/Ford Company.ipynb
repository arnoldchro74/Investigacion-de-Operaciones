# Ford Company

Ford tiene cuatro plantas de automóviles. Cada uno es capaz de
produciendo el Taurus, Lincoln o Escort, pero sólo puede
producir uno de estos coches. El costo fijo de operar cada
planta durante un año y el costo variable de producir un automóvil de
cada tipo en cada planta se dan en la Tabla 51.
Ford enfrenta las siguientes restricciones:
a Cada planta puede producir sólo un tipo de automóvil.
b La producción total de cada tipo de automóvil debe ser de
una sola planta; es decir, por ejemplo, si algún Tauro está
hecho en la planta 1, entonces todos los Tauro deben fabricarse allí.
c Si se utilizan las plantas 3 y 4, entonces la planta 1 también debe
usado.
