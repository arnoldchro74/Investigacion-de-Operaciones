Versión en casete del último álbum de Madonna. una cinta de casete
tiene dos lados (1 y 2). Las canciones de cada cara del casete.
debe tener una duración total de entre 14 y 16 minutos. La longitud
y el tipo de cada canción se dan en la Tabla 28. La tarea
de canciones a la cinta deben cumplir las siguientes condiciones:
1 Cada lado debe tener exactamente dos baladas.
2 La cara 1 debe tener al menos tres canciones exitosas.
3 La canción 5 o la canción 6 deben estar en la cara 1.
4 Si las canciones 2 y 4 están en la cara 1, entonces la canción 5 debe estar
en el lado 2.
