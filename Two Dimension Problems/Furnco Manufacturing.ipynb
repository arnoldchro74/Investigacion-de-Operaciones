# Furnco Manufacturing

Furnco fabrica escritorios y sillas. Cada escritorio utiliza
4 unidades de madera, y cada silla utiliza 3. Un escritorio contribuye
$\$40$ para obtener ganancias y una silla aporta $\$25$. Marketing
Las restricciones requieren que el número de sillas producidas sea de
al menos el doble del número de escritorios producidos. Si 20 unidades de
Hay madera disponible, formule un LP para maximizar el rendimiento de Furnco.
ganancia. Luego resuelve gráficamente el LP.

## Conjuntos


$ i = \text{producto } i$

## Parámetros

$ U_{i} = \textrm{Utilidades del producto } i$

$ M_{i} = \textrm{Materia prima necesitada}\textrm{ para el producto tipo }i$

$ W = \textrm{Materia prima máxima }$

## Variables de decisión

$ x_{i} = \textrm{Cantidad de productos } i \textrm{ producidos}$

## Función objetivo

$$ \textrm{Max } z = \sum_{i = 1}^{n}x_{i}\: U_{i} $$

## Restricciones

No se debe sobrepasar la materia prima $j$,

$$ \sum_{i = 1}^{n}x_{i} M_{i} \leq   W$$

El número de sillas debe ser el doble que los escritorios de madera,

$$ 2x_{1} \leq x_{2} $$

```python
#Se importa la libreria
from pulp import *
```

```python
#Conjuntos
productos = range(2)
```

```python
#Parámetros
U = [40, 25]

M = [4, 3]

W = 20
```

```python
# Crear la instancia del modelo
model = LpProblem(name="Furnco_Manufacturing", sense=LpMaximize)

# Inicialización de las variables de decisión
X = LpVariable.dicts("X", (productos), 0, None, LpContinuous)

# Agregar la función objetivo
model += lpSum([X[i] * U[i] for i in productos])

# Agregar las restricciones

#No se puede rebasar la materia prima
model += lpSum([X[i] * M[i] for i in productos]) <= W

#El número de sillas debe ser el doble que los escritorios de madera
model += 2 * X[0] <= X[1]

# Resolver el problema
model.solve()

# Imprimir el estado del problema
print("Status:", LpStatus[model.status])

# Imprimir las variables
for v in model.variables():
    print(v.name, "=", v.varValue)

# Imprimir la función objetivo
print("Utilidades totales = $", value(model.objective))
```
