# Leary Chemical

Leary Chemical fabrica tres productos químicos: A, B,
y C. Estos productos químicos se producen mediante dos procesos de producción.
procesos: 1 y 2. Ejecutar el proceso 1 durante una hora cuesta $\$4$
y produce 3 unidades de A, 1 de B y 1 de C. Proceso en ejecución
2 por una hora cuesta $\$1$ y produce 1 unidad de A y 1 de B.
Para satisfacer las demandas de los clientes, al menos 10 unidades de A, 5 de B,
y 3 de C deben producirse diariamente. Determinar gráficamente
un plan de producción diario que minimice el costo de cumplir
Las exigencias diarias de Leary Chemical.

## Conjuntos


$ i = \text{producto químico } i$

$ j = \text{Tipo de proceso }j$



## Parámetros

$ C_{j} = \textrm{Costo que genera producir productos A, B y C } $

$ P_{ij} = \textrm{Producción unidades }i  \textrm{ del proceso } j$

$ D_{i} = \textrm{Demanda del producto químico }  i$

## Variables de decisión

$ x_{j} = \textrm{Cantidad de procesos } j \textrm{ llevados a cabo}$

## Función objetivo

$$ \textrm{Min } z = \sum_{j = 1}^{n}x_{j}\: C_{j} $$

## Restricciones

Se debe satisfacer la demanda $j$

$$ \sum_{j = 1}^{n}x_{j} \:P_{ij} \geq  D_{i}, \forall i $$

```python
#Se importa la libreria
from pulp import *
```

```python
#Conjuntos
procesos = range(2)
productos = range(3)
```

```python
#Parámetros
P = [[3, 1, 1], [4, 1, 0]]

C = [4, 1]

D = [10, 5, 3]
```

```python
# Crear la instancia del modelo
model = LpProblem(name="Leary_Chemical", sense=LpMinimize)

# Inicialización de las variables de decisión
X = LpVariable.dicts("X", procesos, 0, None, LpContinuous)

# Agregar la función objetivo
model += lpSum([X[j] * C[j] for j in procesos])

# Agregar las restricciones
for i in productos:
    model += lpSum([X[j] * P[j][i] for j in procesos]) >= D[i]

# Resolver el problema
model.solve()

# Imprimir el estado del problema
print("Status:", LpStatus[model.status])

# Imprimir las variables
for v in model.variables():
    print(v.name, "=", v.varValue)

# Imprimir la función objetivo
print("Costo total = $", value(model.objective))
```
