# Boris Milkem

El gestor de dinero Boris Milkem trata con los franceses
moneda (el franco) y moneda americana (el dólar). En
12 de la noche, puede comprar francos pagando 0,25 dólares por
franco y dólares pagando 3 francos por dólar. sea ​​x1
número de dólares comprados (pagando francos) y x2
número de francos comprados (pagando dólares). Asumir que
ambos tipos de transacciones se realizan simultáneamente, y
la única restricción es que a las 00:01 a.m. Boris debe tener un
número no negativo de francos y dólares.
a Formule un PL que permita a Boris maximizar
la cantidad de dólares que tiene después de realizar todas las transacciones
terminado

## Conjuntos

$ i = \text{Moneda tipo }i$

## Parámetros

$ U_{i} = \textrm{Utilidad de la moneda } i$

## Variables de decisión

$ x_{1} = \textrm{unidades monetarias de francos a convertir en dólares } $

$ x_{2} = \textrm{unidades monetarias de dólares a convertir en francos } $

## Función objetivo

$$ \textrm{Max } z = x_{1} - 0.25x_{2}  $$

## Restricciones


$$  x_{1} - 0.25x_{2} \geq 0 $$

$$   3 x_{2} - x_{1} \geq 0 $$



```python
#Se instala la libreria
!pip install pulp
```

```python
#Se importa la libreria
from pulp import *
```

```python
model = LpProblem(name = "Boris_Milkem", sense = LpMaximize)

#Inicialización de las variables de decisión
X1 = LpVariable('X1', 0, None, LpInteger)
X2 = LpVariable('X2', 0, None, LpInteger)

model += X1 - 0.25 * X2

#Restricciones

model += X1 - 0.25 * X2 >= 0
model += 3 * X2 - X1 >= 0

#Se resuelve el modelo
model.solve()

#Se imprime el status del problema
print("Status", LpStatus[model.status])

#Se imprimen las variables
for v in model.variables():
  print(v.name, "=", v.varValue)

print("Función Objetivo =", value(model.objective))
```
