# Eli Daisy

Eli Daisy utiliza las sustancias químicas 1 y 2 para producir dos medicamentos.
El fármaco 1 debe contener al menos un 70 % de la sustancia química 1 y el fármaco 2 debe contener
tener al menos un 60 % de sustancia química 2. Se pueden consumir hasta 40 oz de medicamento 1.
se vende a 6 dólares la onza; Se pueden vender hasta 30 onzas del medicamento 2 a $\$5$.
por onza. Se pueden comprar hasta 45 oz del químico 1 a $\$6$
por oz, y se pueden comprar hasta 40 oz de producto químico 2 en
$4 por onza. Formule un LP que pueda usarse para maximizar
Las ganancias de Daisy.

#Conjuntos

$ i = \text{químico indexado por } i$

$ j = \text{Medicamento indexado por } j$

# Parámetros

$U_{ij} = \textrm{Utilidad de vender el fertilizante } j \textrm{ utilizando el químico } i$

$ MP_{i} = \textrm{Materia prima indexada por } i $

$ F_{j} = \textrm{Cantidad máxima de medicamento } j $

## Variables de decisión

$ x_{ij} = \textrm{Cantidad producida de medicamento }j \textrm{ con el químico }i$

## Función Objetivo

$$ \textrm{Max } z = \sum_{j = 1}^{n}\sum_{i = 1}^{n} U_{ij}\: x_{ij} $$

No se puede rebasar la cantidad de stock en inventario,

$$ \sum_{j = 1}^{n} x_{ij} \leq MP_{i}, \forall i$$

Hay una cantidad limitada de medicamento,

$$ \sum_{i = 1}^{n} x_{ij} \leq F_{j}, \forall j$$

El medicamento uno debe tener al menos 70% de la sustancia química 1,

$$ x_{11} \geq 0.7 \sum_{i = 1}^{n} x_{ij}$$

El medicamento dos debe tener al menos 60% de la sustancia química 2,

$$ x_{22} \geq 0.6 \sum_{i = 1}^{n} x_{ij}$$

```python
#Se importa la libreria
from pulp import *
```

```python
#Conjuntos
quimico = range(2)
medicamento = range(2)
```

```python
#Parámetros

U = [[0, -1], [2, 1]]

MP = [45, 40]

F = [40, 30]
```

```python
# Crear la instancia del modelo
model = LpProblem(name="Eli_Daisy", sense=LpMaximize)

# Inicialización de las variables de decisión
X = LpVariable.dicts("X", (quimico, medicamento), 0, None, LpContinuous)

# Agregar la función objetivo
model += lpSum([[U[i][j] * X[i][j] for i in quimico] for j in medicamento])

# Agregar las restricciones

for i in quimico:
  model += lpSum([X[i][j] for j in medicamento]) <= MP[i]

for j in medicamento:
  model += lpSum([X[i][j] for i in quimico]) <= F[j]

model += X[0][0] >= 0.7 * (X[0][0] + X[1][0])
model += X[1][1] >= 0.6 * (X[0][1] + X[1][1])

# Resolver el problema
model.solve()

# Imprimir el estado del problema
print("Status:", LpStatus[model.status])

# Imprimir las variables
for v in model.variables():
    print(v.name, "=", v.varValue)

# Imprimir la función objetivo
print("Utilidades totales = $", value(model.objective))
```
