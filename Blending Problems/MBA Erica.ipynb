# MBA Erica

La joven MBA Erica Cudahy puede invertir hasta 1.000 dólares. Ella
puede invertir su dinero en acciones y préstamos. cada dolar
invertido en acciones produce una ganancia de 10 centavos y cada dólar invertido
en un préstamo produce una ganancia de 15 centavos. Al menos el 30% de todo el dinero invertido.
deben estar en acciones y al menos $400 deben estar en préstamos.
Formule un LP que pueda usarse para maximizar la ganancia total
obtenido de la inversión de Erica. Luego resuelve gráficamente
el LP.

#Conjuntos

$ i = \text{Inversiones indexado por } i$

# Parámetros

$I_{i} = \textrm{Ingreso por lo invertido } i $

$ P = \textrm{Presupuesto para la inversión } $

## Variables de decisión

$ x_{i} = \textrm{Cantidad invertida }i $

## Función Objetivo

$$ \textrm{Max } z = \sum_{i = 1}^{n} I_{i}\: x_{i} $$

No se puede rebasar el presupuesto,

$$ \sum_{I = 1}^{n} x_{i} \leq P $$

Al menos el 30% de todo el dinero invertido debe estar en acciones,

$$ x_{1} \geq  0.3 \sum_{I = 1}^{n} x_{i} $$

Al menos se debe invertir $400 en préstamos,

$$ x_{2} \geq  400 $$

```python
#Se importa la libreria
from pulp import *
```

```python
#Conjuntos
inversiones = range(2)
```

```python
#Parámetros

I = [0.1, 0.15]

P = 1000
```

```python
# Crear la instancia del modelo
model = LpProblem(name="MBA_Erica", sense=LpMaximize)

# Inicialización de las variables de decisión
X = LpVariable.dicts("X", (inversiones), 0, None, LpContinuous)

# Agregar la función objetivo
model += lpSum([I[i] * X[i] for i in inversiones])

# Agregar las restricciones

model += lpSum([X[i] for i in inversiones]) <= P

model += X[0] >= 0.3 * lpSum([X[i] for i in inversiones])
model += X[1] >= 400

# Resolver el problema
model.solve()

# Imprimir el estado del problema
print("Status:", LpStatus[model.status])

# Imprimir las variables
for v in model.variables():
    print(v.name, "=", v.varValue)

# Imprimir la función objetivo
print("Utilidades totales = $", value(model.objective))
```
