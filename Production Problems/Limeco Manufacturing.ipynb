# LIMECO Manufacturing

LIMECO posee una fábrica de cal y vende seis grados de
cal (grados 1 a 6). El precio de venta por libra es
figura en el Cuadro 25. La cal se produce en hornos. Si se ejecuta un horno
para un turno de 8 horas, se producen las cantidades (en libras) de cada grado de cal dadas en la Tabla 26. Cuesta $\$150$ ejecutarlo
un horno para un turno de 8 horas. Cada día la fábrica lo cree
puede vender hasta las cantidades (en libras) de cal dadas en
Tabla 27.
La cal producida por el horno puede ser reprocesada por
utilizando cualquiera de los cinco procesos descritos en la Tabla 28.
Por ejemplo, a un costo de $1/lb, una libra de cal grado 4
puede transformarse en 0,5 lb de cal grado 5 y 0,5 lb de
cal grado 6.
Cualquier sobrante de lima al final de cada día debe ser
eliminado, con los costos de eliminación (por libra) indicados en
Tabla 29.
Formule un LP cuya solución le indique a LIMECO cómo
para maximizar sus ganancias diarias.
