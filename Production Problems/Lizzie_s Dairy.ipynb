#Conjuntos

$ i = \text{Leche alta o baja en calorías indexado por } i$

$ j = \text{Cantidad de producto indexado por } j$


## Variables de decisión

$ x_{ij} = \textrm{Cantidad producida de producto }j \textrm{ con la leche }i$

$ c_{i} = \textrm{cantidad de crema utilizada para la producción } i$


## Función Objetivo

$$ \textrm{Ganancias }= 1.5(x_{11}+x_{21}+c_{1}) + 1.2(0.9x_{12} + 0.9x_{22} + 0.9 c_{2}) $$

$$ \textrm{Costos }= 0.8(x_{11}+x_{12}+x_{13}) + 0.4(x_{21} + x_{22} + x_{23}) + 0.4(x_{13} + x_{23}) + 0.4(x_{11} + x_{21} + c_{1}) + 0.4(0.9x_{12} + 0.9x_{22} + 0.9c_{2})$$

$$ \textrm{Max } \textrm{Utilidad }= \textrm{Ganancias } - \textrm{Costos }$$

## Restricciones

La máquina tiene una capacidad de 3000 lb,

$$ x_{11} + x_{12} + c_{1} + x_{21} + x_{22} + c_{2} \leq 3000 $$

Lo mínimo y máximo que puede producir la empresa sobre queso crema,

$$ x_{11} + x_{21} + c_{1} \geq 1000$$

$$ x_{11} + x_{21} + c_{1} \leq 1500$$

Lo mínimo y máximo que puede producir la empresa sobre requesón,

$$ 0.9x_{12} + 0.9x_{12} + 0.9c_{2} \geq 1000$$

$$ 0.9x_{12} + 0.9x_{12} + 0.9c_{2} \leq 2000$$

Lo máximo de leche evaporada que se tiene,

$$ x_{13} + x_{23} \leq 2000 $$

Se debe tener un peso específico para la leche,

$$ 0.6x_{11} + 0.3x_{21} \geq 0.5 (x_{11} + x_{21}) $$

Se debe tener un peso específico para el requesón,

$$ 0.6x_{12} + 0.3x_{22} \geq 0.35 (x_{12} + x_{22}) $$

Se debe tener un peso específico para la nata en el queso crema,

$$ c_{1} \geq 0.4(x_{11} + x_{21} + c_{1}) $$

Se debe tener un peso específico para la nata en el requesón,

$$ c_{2} \geq 0.2(x_{12} + x_{22} + c_{2}) $$

Se produce crema evaporando,

$$ 0.6x_{13} + 0.3x_{23} \geq c_{1} + c_{2} $$

```python
from re import X
# Crear la instancia del modelo
model = LpProblem(name="ABC_Company", sense=LpMaximize)

X11 = LpVariable('X11', 0, None, LpContinuous)
X12 = LpVariable('X12', 0, None, LpContinuous)
X13 = LpVariable('X13', 0, None, LpContinuous)
X21 = LpVariable('X21', 0, None, LpContinuous)
X22 = LpVariable('X22', 0, None, LpContinuous)
X23 = LpVariable('X23', 0, None, LpContinuous)
C1 = LpVariable('C1', 0, None, LpContinuous)
C2 = LpVariable('C2', 0, None, LpContinuous)

# Agregar la función objetivo
model += 1.5 * (X11 + X21 + C1) + 1.2 * (0.9 * X12 + 0.9 * X22 + 0.9 * C2) - 0.8 * (X11 + X12 + X13) - 0.4 * (X21 + X22 + X23) - 0.4 * (X13 + X23) - 0.4 * (X11 + X21 + C1) - 0.4 * (0.9 * X12 + 0.9 * X22 + 0.9 * C2)

# Agregar las restricciones
model += X11 + X12 + C1 + X21 + X22 + C2 <= 3000
model += X11 + X21 + C1 >= 1000
model += X11 + X21 + C1 <= 1500
model += 0.9 * X12 + 0.9 * X22 + 0.9 * C2 >= 1000
model += 0.9 * X12 + 0.9 * X22 + 0.9 * C2 <= 2000
model += X13 + X23 <= 2000
model += 0.6 * X11 + 0.3 * X21 >= 0.5 * (X11 + X21)
model += 0.6 * X12 + 0.3 * X22 >= 0.35 * (X12 + X22)
model += C1 >= 0.4 * (X11 + X21 + C1)
model += C2 >= 0.2 * (X12 + X22 + C2)
model += 0.6 * X13 + 0.3 * X23 >= C1 + C2

# Resolver el modelo
model.solve()

# Imprimir el estado del problema
print("Status:", LpStatus[model.status])

# Imprimir las variables
for v in model.variables():
    print(v.name, "=", v.varValue)

# Imprimir la función objetivo
print("Utilidades totales = $", value(model.objective))
```
