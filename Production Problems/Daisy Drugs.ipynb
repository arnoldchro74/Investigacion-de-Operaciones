# Daisy Drugs

Daisy Drugs fabrica dos medicamentos: 1 y 2. El
Las drogas se producen mezclando dos sustancias químicas: 1 y 2. Por
peso, el medicamento 1 debe contener al menos un 65 % del químico 1, y
El medicamento 2 debe contener al menos un 55 % de la sustancia química 1. El medicamento 1 se vende
por $6/oz$, y el medicamento 2 se vende por $4/oz$. Los productos químicos 1 y 2 pueden
ser producido mediante uno de dos procesos de producción. Correr
El proceso 1 durante una hora requiere 3 oz de materia prima y 2
horas de mano de obra calificada y rinde 3 oz de cada químico.
Ejecutar el proceso 2 durante una hora requiere 2 oz de materia prima
y 3 horas de mano de obra calificada y rinde 3 oz de químico 1
y 1 oz de químico 2. Un total de 120 horas de mano de obra calificada
y se encuentran disponibles 100 oz de materia prima. Formular un LP
que se puede utilizar para maximizar los ingresos por ventas de Daisy.

##Conjuntos

$ i = \text{Químico indexado por } i$

$ j = \text{Medicamento indexado por } j$

$ k = \text{Proceso indexado por } k$

## Parámetros

$U_{ij} = \textrm{Utilidad de vender el medicamento } j \textrm{ utilizando el químico } i$

$ MP = \textrm{Materia prima total } $

$ P{i} = \textrm{Materia prima indexada por } i $

$L_{ik} = \textrm{Onzas de materia prima } i \textrm{ en el proceso } k$

$ H_{i} = \textrm{Horas de trabajo indexada por } i $

$ HT = \textrm{Horas de trabajo totales } $

## Variables de decisión

$ x_{ij} = \textrm{Cantidad producida de medicamento }j \textrm{ con el químico }i$

$ t_{i} = \textrm{Materia prima indexado por } i $

## Función Objetivo

$$ \textrm{Max } z = \sum_{j = 1}^{n}\sum_{i = 1}^{n} U_{ij}\: x_{ij} $$

El medicamento 1 debe contener al menos por peso un 65% del químico 1,

$$ x_{11} \geq 0.65 \sum_{i = 1}^{n} x_{i1}  $$

El medicamento 2 debe contener al menos por peso un 55% del químico 1,

$$ x_{12} \geq 0.55 \sum_{i = 1}^{n} x_{i2}  $$

Se recibe oz de cada materia prima por cada hora en el proceso 1 y 2,

$$ \sum_{i = 1}^{n} t_{i} L_{ik} \geq \sum_{i = 1}^{n} x_{ik}, \forall k $$

No se puede rebasar la cantidad de stock en inventario de materia prima,

$$ \sum_{i = 1}^{n} t_{i} P_{i} \leq MP$$

No se puede rebasar el número de horas de mano de obra,

$$ \sum_{i = 1}^{n} t_{i} H_{i} \leq HT$$

```python
#Se importa la libreria
from pulp import *
```

```python
#Conjuntos
químico = range(2)
medicamento = range(2)
proceso = range(2)
```

```python
#Parámetros

U = [[6, 6], [4, 4]]

MP = 100

P = [3, 2]

L = [[3, 3], [3, 1]]

H = [2, 3]

HT = 120
```

```python
# Crear la instancia del modelo
model = LpProblem(name="Daisy_Drugs", sense=LpMaximize)

# Inicialización de las variables de decisión
X = LpVariable.dicts("X", (químico, medicamento), 0, None, LpContinuous)
T = LpVariable.dicts("T", químico, 0, None, LpContinuous)

# Agregar la función objetivo
model += lpSum([[U[j][i] * X[i][j] for i in químico] for j in medicamento])

# Agregar las restricciones

model += X[0][0] >= 0.65 * (X[0][0] + X[1][0])
model += X[0][1] >= 0.55 * (X[0][1] + X[1][1])

for k in proceso:
  model += lpSum([T[i] * L[i][k] for i in químico]) >= lpSum([X[i][k] for i in químico])

model += lpSum([P[i] * T[i] for i in químico]) <= MP

model += lpSum([H[i] * T[i] for i in químico]) <= HT

# Resolver el problema
model.solve()

# Imprimir el estado del problema
print("Status:", LpStatus[model.status])

# Imprimir las variables
for v in model.variables():
    print(v.name, "=", v.varValue)

# Imprimir la función objetivo
print("Utilidades totales = $", value(model.objective))
```
